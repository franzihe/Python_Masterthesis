```python
def plt_scatter_obs_model(merged_mean_SWP, SWP_model, label,
                          intercept, gradient, day):
    mpl.style.use('ggplot')
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)

    ax.scatter(merged_mean_SWP, SWP_model[0][:np.asarray(merged_mean_SWP).shape[0]],
              color = colors[int(day)-20], alpha = 0.7, s=150, label=label)
    for ens_memb in range(1,10):
        ax.scatter(merged_mean_SWP, SWP_model[ens_memb][:np.asarray(merged_mean_SWP).shape[0]],
              color = colors[int(day)-20], alpha = 0.7, s=150, label='_nolegend_')

    ax.plot([-250, 5000], [intercept+gradient*(-250), intercept+gradient*5000],
           color= colors[int(day)-20], linewidth = 3.,
                alpha=0.8,label='_nolegend_')
    ax.text(0.49,0.9, 'y = {:.2f} + {:.2f}x'.format(intercept, gradient),
                        verticalalignment ='top', horizontalalignment='left', color = colors[int(day)-20],
                         transform = ax.transAxes, fontsize = spagh.label_fs-4,
                         bbox={'facecolor':'white','alpha':.8,'pad':10})
    #plt.scatter(x_obs,y_obs)
    #plt.plot(x_obs,intercept+gradient*x_obs)

    lgd = ax.legend(loc='upper right',
             #fancybox=True, shadow=True, #ncol=3, 
              fontsize=spagh.label_fs-8,scatterpoints = 1)
    frame = lgd.get_frame()
    frame.set_facecolor('white')
    plt.setp(lgd.get_texts(), color=spagh.blue)

    ax.plot([-500,5500], [-500,5500], linestyle='-',color=spagh.memb_col)
    ## yaxis
    ax.set_ylim(-250,5500)
    ax.set_yticks(np.arange(0,6000,500))
    ax.set_yticklabels([0,'',1000,'',2000,'',3000,'',4000,'',5000,''],fontsize = spagh.tick_fs)

    # xaxis
    ax.set_xlim(-250,5500)
    ax.set_xticks(np.arange(0,6000,500))
    ax.set_xticklabels([0,'',1000,'',2000,'',3000,'',4000,'',5000,''],fontsize = spagh.tick_fs)

    ax.set_ylabel('MEPS forecast', fontsize=spagh.label_fs)
    ax.set_xlabel('observation', fontsize=spagh.label_fs)

    plt.tight_layout()
```
